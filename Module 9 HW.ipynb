```
#max, min, average, standard deviation, geometric average functions
theList = [2, 5, 7, 10, 12, 15, 18, 21, 25, 28, 45 ,82, 16, 30, 45, 62]

def theMax(numbers):
  max = numbers[0]
  max = -1
  for k in numbers:
    if max < k:
      k = max
    return max

def theMin(numbers):
  min = 10000
  for k in numbers:
    if min > k:
      min = k
  return min

def theAverage(numbers):
  total = 0
  for k in numbers:
    total += k
  return total/ len(numbers)

def theSD(numbers):
  total = 0
  avg = theAverage(numbers)
  for k in numbers:
    total += pow(k-avg,2)
  return pow(total/(len(numbers)-1),.5)

def theGeoMean(numbers):
  total = 1
  for k in numbers:
    total *= k
  return pow(total, 1/len(numbers))

theList = [19, 32, 34, 33, 75, 91, 9, 4, 18, 38]
print(theAverage(theList))
print(theGeoMean(theList))
```

```
#sum of even valued fibonacci numbers
def fib():
  a = 1
  b = 1
  total = 0

  for k in range(1000):
    c = a + b
    if c >= 1000:
        break
    elif (c % 2 == 0):
      total += c 
    a = b
    b = c

  return total

print(fib())
```
